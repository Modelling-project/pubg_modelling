## 머신러닝 점수 체크(최종 단계)

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
k = 0
for i in [Dsquad_fpp, Dduo_fpp, Dsquad, Dsolo_fpp, Dduo,
          Dsolo, Dnormal_squad_fpp, Dcrashfpp, Dnormal_duo_fpp,
         Dflaretpp, Dnormal_solo_fpp, Dflarefpp, Dnormal_squad,
         Dcrashtpp, Dnormal_solo, Dnormal_duo]:
    j = ['Dsquad_fpp', 'Dduo_fpp', 'Dsquad', 'Dsolo_fpp', 'Dduo',
          'Dsolo', 'Dnormal_squad_fpp', 'Dcrashfpp', 'Dnormal_duo_fpp',
         'Dflaretpp', 'Dnormal_solo_fpp', 'Dflarefpp', 'Dnormal_squad',
         'Dcrashtpp', 'Dnormal_solo', 'Dnormal_duo']
    
    feature = i[['boosts', 'damageDealt', 'headshotKills', 'heals',
       'killPlace', 'kills', 'killStreaks', 'longestKill',
       'rideDistance', 'swimDistance', 'vehicleDestroys',
       'walkDistance', 'weaponsAcquired']]
    target = i["winPlacePerc"]
    
    train_x, test_x, train_y, test_y = train_test_split(feature, target, test_size=0.3, random_state=0)
    
    train_x, X_val, train_y, y_val = train_test_split(train_x, train_y, test_size=0.25, random_state=0)
    
    globals()['reg_1{}'.format(i)] = Lasso()
    globals()['reg_2{}'.format(i)] = Ridge()

    globals()['reg_1{}'.format(i)].fit(train_x, train_y)
    globals()['reg_2{}'.format(i)].fit(train_x, train_y)

    pred_train1 = globals()['reg_1{}'.format(i)].predict(train_x)
    pred_train2 = globals()['reg_2{}'.format(i)].predict(train_x)

    pred_val1 = globals()['reg_1{}'.format(i)].predict(X_val)
    pred_val2 = globals()['reg_2{}'.format(i)].predict(X_val)

    mse_train1 = mean_squared_error(train_y, pred_train1)
    mse_val1 = mean_squared_error(y_val, pred_val1)
    mse_train2 = mean_squared_error(train_y, pred_train2)
    mse_val2 = mean_squared_error(y_val, pred_val2)

    print(f'====================={j[k]}=====================')
    print("1. Lasso\t, train=%.4f, val=%.4f" % (mse_train1, mse_val1))
    print("2. Ridge\t, train=%.4f, val=%.4f" % (mse_train2, mse_val2))
    print('')
    k += 1
```
